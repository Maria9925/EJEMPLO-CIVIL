#### Question 13 (***)

A rigid cylinder with inside diameter of 15 mm contains a column of water 500 mm long. What will the column length be if a force of 2 kN is applied to its top by a frictionless plunger? Assume a bulk modulus of 2.21 GPa.

```{dropdown} Answer

$497 \; mm$
```

```{dropdown} Hint
The compressibility of liquids is described by the following equation:  

$$\Delta p =-E_v \frac{\Delta v}{v}$$   

The change in pressure can be calculated using the force and area as:

$$ \Delta p = \frac {\Delta F}{A}$$

The cylinder's cross-sectional area is fixed, but the column length is changing, modifying the fluid's volume

[Click here for a video explaining the problem](https://auckland.au.panopto.com/Panopto/Pages/Viewer.aspx?id=21e9c540-fff5-445e-8313-b294015a2d1e)
```

**Solution**

```{dropdown} Show code cell outputs
The compressibility of liquids is described by the following equation:  

$$\Delta p =-E_v \frac{\Delta v}{v}\\
$$   
where:

- $\Delta p$ is the change in pressure 
- $E_v$ is the bulk modulus of elasticity 
- $\Delta v$ is the change in the specific volume 
- $v$ is the original specific volume

The change in pressure can be calculated using the force and area as:

$$\Delta p = \frac {\Delta F}{A}$$

Changing specific volume with the following equation, we have 

$$\frac {\Delta v }{v}= \frac {\frac{\Delta Vol}{m}}{\frac{Vol}{m}}$$

Knowing that the change in volume is due to the change in length since the area remains constant, we have the following:

$$
\frac {\Delta v }{v}= \frac{L_2-L_1}{L_1}
$$

Substituting into the compressibility equation, we have:

$$\Delta p =-E_v \frac{L_2-L_1}{L_1}$$  

Solving for $L_2$:  

$$L_2 = -L_1 \cdot \frac{\Delta p}{E_v}+ L_1$$  
```

````python
%%render long 4
dF=2000 #N
d=0.015 #m
L_1=0.5 #m
E_v= 2.21*10**9 #Pa
A= pi*(d/2)**2 #m2
dp= dF/A #Pa
L_2= -L_1*dp/E_v+L_1 #m
L_2_=L_2*1000 #mm
````

#### Question 7 (***)

A journal bearing consists of an 80.0 mm diameter shaft in an 80.4 mm sleeve 120 mm long. The clearance space is filled with oil with a viscosity of 0.11 N.s/m<sup>2</sup>. Calculate the torque required to keep the shaft turning at 150 rpm.

```{dropdown} Answer

$0.417 \; N.m$
```

```{dropdown} Hint
The torque is calculated by multiplying the force with the distance

$$
T= F \cdot r_{shaft}\\
$$

where the force can be calculated from viscosity.  

[Click here for a video explaining the problem](https://auckland.au.panopto.com/Panopto/Pages/Viewer.aspx?id=ef2dfea1-f7b0-40d9-9b20-b29d01756b86)

```

````python
from PIL import Image
img = Image.open(r"figures/4_viscosity_7.png")
base_width = 350 # Change this to your desired width
w_percent = base_width / float(img.size[0])
h_size = int(float(img.size[1]) * w_percent)
resized_img = img.resize((base_width, h_size))
resized_img
````

**Solution**

```{dropdown} Show code cell outputs


Torque:

$$
T= F \cdot r_{shaft}\\
$$

With:

$$
F=\tau A\\
$$

and

$$
\tau= \mu\frac {\Delta u}{\Delta y}
$$

Assuming linear velocity distribution

$$
u= wr
$$

$$
w= \frac {2 \pi N}{60}
$$

$$
u=\frac{r_{shaft}\cdot2\cdot\pi\cdot N_{rpm}}{60}
$$

```

````python
%%render long 4
d_shaft= 80/1000 #m
d_sleeve= 80.4/1000 #m
L_shaft= 120/1000 #m
mu=0.11 #Ns/m2
N_rpm=150 #rpm

r_shaft= (d_shaft)/2 #m

u= (r_shaft*2*pi*N_rpm)/60 #m/s

Delta_y= ((d_sleeve-d_shaft))/2 #m

tau= mu*u/Delta_y #Pa

F= tau*pi*(d_shaft)*(L_shaft) #N

T= F*r_shaft #N.m
````
